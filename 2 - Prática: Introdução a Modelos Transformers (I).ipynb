# Perguntas e Respostas

Definição do pipeline

```python
qea_pipeline = pipeline(
    task="question-answering",
    model="pierreguillou/bert-base-cased-squad-v1.1-portuguese"
  )
```

Texto base

```python
texto = 'O time do Fortaleza ganhou do Flamengo por 1 a 0 ontem, mas sofreu para manter esse resultado.'
```

Perguntas

```python
pergunta1 = 'Quem ganhou o jogo ontem?'
resposta1 = qea_pipeline(
    context = texto,
    question = pergunta1
)
print(pergunta1, '\nR:', resposta1['answer'], '\nScore da resposta:', resposta1['score'])
```

```python
pergunta2 = 'Qual foi o resultado do jogo?'
resposta2 = qea_pipeline(
    context = texto,
    question = pergunta2
)
print(pergunta2, '\nR:', resposta2['answer'], '\nScore da resposta:', resposta2['score'])
```

```python
pergunta3 = 'Foi um jogo fácil?'
resposta3 = qea_pipeline(
    context = texto,
    question = pergunta3
)
print(pergunta3, '\nR:', resposta3['answer'], '\nScore da resposta:', resposta3['score'])
```

# Geração de Resumos

Definição do pipeline

```python
res_pipeline = pipeline(
    task="summarization"
  )
```

Texto

```python
texto = '''
O Flamengo sofreu um duro golpe no Brasileirão neste sábado (25). No Castelão, em compromisso válido pela 30ª rodada, o Rubro-Negro visitou o Fortaleza e foi derrotado por 1 a 0, com gol marcado pelo ex-Palmeiras Breno Lopes.

Com o resultado, o Flamengo segue com 61 pontos somados e empatado com o Palmeiras, mas pode ver o rival voltar a abrir vantagem na liderança em caso de vitória ou empate com o Cruzeiro neste domingo (26).
ESPN no Disney+

Assista à ESPN no Disney+ Esportes da ESPN, filmes e séries em um só lugar. Encontre tudo que você gosta no DisneyPlus.

Prestes a disputar uma vaga na final da CONMEBOL Libertadores com o Racing, na próxima quarta-feira (29), em Avellaneda, o Flamengo preservou alguns jogadores para a partida, incluindo Léo Ortiz, Alex Sandro e De La Cruz.

No lado do Laion, o resultado foi importante para a equipe, que foi aos 27 pontos e ganhou um gás extra na luta contra o rebaixamento, apesar de seguir no Z-4.

O único gol marcado saiu com 12 minutos do primeiro tempo. Após contra-ataque puxado pelo time da casa, Breno Lopes chutou forte, no canto de Rossi, e fez 1 a 0. O atacante chegou a estufar as redes uma segunda vez, mas estava em posição irregular.
Relacionado

No segundo tempo, o Flamengo teve chances de empatar a partida, mas não teve sorte na pontaria, ao mesmo tempo que o Fortaleza se segurou muito bem na defesa.

O Rubro-Negro, mais de uma vez, ficou muito perto de estufar as redes, dominando as principais ações em campo.

E mesmo com as mudanças feitas pelo técnico Filipe Luís, os cariocas não conseguiram uma reação em solo cearense.
'''
```

```python
resumo = res_pipeline(texto, max_length=100, min_length=50)
print(resumo)
```

# Geração de Texto

Definição do Pipeline

```python
ger_pipeline = pipeline(
    task = "text-generation",
    model = "pierreguillou/gpt2-small-portuguese"
)
```

Texto referência

```python
texto = 'O Fortaleza é um time bom, mas está com problemas para ganhar jogos.'
```

Texto gerado

```python
resultado = ger_pipeline(texto, max_length=10, do_sample=True)
print(resultado)
```
